```python
from ipywidgets import widgets,interactive,interactive_output
import pandas as pd
import glob
import os
import numpy as np
```

```python
# setting data directories
DATA_DIR = os.getcwd()+'/Data'
SIG_DIR = os.getcwd()+'/Data/SignalFiles'
CONFIG_DIR = os.getcwd()+'/Configurations'
REGIME_DIR = os.getcwd()+'/Data/RegimeFiles'
ORDER_DF_REF = pd.DataFrame()
SIGNAL_NAMES = []
```

```python
# retrieving file names 
all_files = glob.glob(DATA_DIR +"/*.csv")
all_files = [n.split('/')[-1] for n in all_files]

# retrieving signal file names 
all_signal_files = glob.glob(SIG_DIR +"/*.csv")
all_signal_files = [n.split('/')[-1] for n in all_signal_files]

# retrieving available files
signalfiles = [n.split('.csv')[0] for n in all_signal_files]

# retrieving available tickers and timeframes
tickers = [n.split('_')[0] for n in all_files]
tickers = set(tickers)
frequency = [n.split('_')[1].split('.')[0] for n in all_files]
frequency = set(frequency)

# creating dropdowns - data
t = widgets.Dropdown(description='Ticker',options=tickers)
f = widgets.Dropdown(description='Frequency',options=frequency)
# creating dropdowns - signal data
s = widgets.Dropdown(description='Signal File',options=signalfiles)


# creating dropdowns - date options 
start = widgets.DatePicker(description='Start Date')
end = widgets.DatePicker(description='End Date')

# results button
submit_data = widgets.Button(description='Upload Data')
submit_signal_data = widgets.Button(description='Upload Signal Data')

# defining data retrieval functions
def get_data(t,f,start,end):
    filename = f'{t}_{f}.csv'
    if filename in all_files:
        df = pd.read_csv(DATA_DIR+f'/{filename}',index_col=0, parse_dates=[['Date','Time']])
        df = df.dropna()
        df.index = pd.to_datetime(df.index)
        try:
            df.loc[(df.index.date>=start) & (df.index.date<=end)]
            display(df.head(5))
            return df
        except Exception as e:
            print(f'Error {e}')
    else:
        print('File doesnt exist')

def get_signal_data(s):
    global SIGNAL_NAMES
    filename = f'{s}.csv'
    df = pd.read_csv(SIG_DIR+f'/{filename}',index_col=0, parse_dates=[['Date','Time']]).dropna()
    display(df.head(5))
    SIGNAL_NAMES = df.columns
    return df
    
# CREATING INTERACTIVE BUTTONS 
DATA_DF = interactive(get_data,{'manual': True, 'manual_name':'Upload Data'},t=t,f=f,start=start,end=end)
SGNL_DF = interactive(get_signal_data,{'manual': True, 'manual_name':'Upload Signal Data'},s=s)

# displaying
DATA_UI = widgets.Accordion(children=[DATA_DF,SGNL_DF])
DATA_UI.set_title(0, 'Historical Data')
DATA_UI.set_title(1, 'Signal Data')
```

```python
# creating reference variables
dir_options = ['LONG','SHORT']
order_options = ['MARKET','LIMIT']
signal_names = ['','','','']

if len(SIGNAL_NAMES)>0:
    signal_names = SIGNAL_NAMES
    

# creating UI
l1 = widgets.HTML(value=f'<b>Signal 1: {signal_names[0]}</b>')
lo1 = widgets.HTML(value=f'<b>Signal 1: {signal_names[0]}</b>')
d1 = widgets.Dropdown(description='Direction',options=dir_options)
o1 = widgets.Dropdown(description='Order Type',options=order_options)
li1 = widgets.FloatText(description='Limit Price')
pt1 = widgets.FloatText(description='Profit Target')
pt1_opt = widgets.FloatRangeSlider(description='PT OPT',min=0,max=10.0,step=0.1)
pt1_step = widgets.FloatText(description='Step', value=1)
sl1 = widgets.FloatText(description='Stop Loss')
sl1_opt = widgets.FloatRangeSlider(description='SL OPT',min=0,max=10.0,step=0.1)
sl1_step = widgets.FloatText(description='Step', value=1)
c1 = widgets.FloatText(description='Cancel > N Bars')
c1_opt = widgets.FloatRangeSlider(description='Cancel OPT',min=0,max=10.0,step=0.1)
c1_step = widgets.FloatText(description='Step', value=1)
h1 = widgets.FloatText(description='Max Hold Bars')
h1_opt = widgets.FloatRangeSlider(description='Hold Opt',min=0,max=10.0,step=0.1)
h1_step = widgets.FloatText(description='Step', value=1)
q1 = widgets.FloatText(description='Quantity: ')
r1 = widgets.Checkbox(value=False, description='Reduce only')
v1 = widgets.Checkbox(value=False, description='Adjust for Volatility')
f1 = widgets.Checkbox(value=False, description='Force EOD')
opt1 = widgets.Checkbox(value=False,description='Run Optimized Backtest')
signal1_ui = widgets.VBox([l1,d1,o1,pt1,sl1,c1,h1,q1,r1,v1,f1])


##########


l2 = widgets.HTML(value=f'<b>Signal 2: {signal_names[1]}</b>')
lo2 = widgets.HTML(value=f'<b>Signal 2: {signal_names[1]}</b>')
d2 = widgets.Dropdown(description='Direction',options=dir_options)
o2 = widgets.Dropdown(description='Order Type',options=order_options)
li2 = widgets.FloatText(description='Limit Price')
pt2 = widgets.FloatText(description='Profit Target')
pt2_opt = widgets.FloatRangeSlider(description='PT OPT',min=0,max=10.0,step=0.1)
pt2_step = widgets.FloatText(description='Step', value=1)
sl2 = widgets.FloatText(description='Stop Loss')
sl2_opt = widgets.FloatRangeSlider(description='SL OPT',min=0,max=10.0,step=0.1)
sl2_step = widgets.FloatText(description='Step', value=1)
c2 = widgets.FloatText(description='Cancel > N Bars')
c2_opt = widgets.FloatRangeSlider(description='Cancel OPT',min=0,max=10.0,step=0.1)
c2_step = widgets.FloatText(description='Step', value=1)
h2 = widgets.FloatText(description='Max Hold Bars')
h2_opt = widgets.FloatRangeSlider(description='Hold Opt',min=0,max=10.0,step=0.1)
h2_step = widgets.FloatText(description='Step', value=1)
q2 = widgets.FloatText(description='Quantity: ')
r2 = widgets.Checkbox(value=False, description='Reduce only')
v2 = widgets.Checkbox(value=False, description='Adjust for Volatility')
f2 = widgets.Checkbox(value=False, description='Force EOD')
opt2 = widgets.Checkbox(value=False,description='Run Optimized Backtest')
signal2_ui = widgets.VBox([l2,d2,o2,pt2,sl2,c2,h2,q2,r2,v2,f2])

#########


l3 = widgets.HTML(value=f'<b>Signal 3: {signal_names[2]}</b>')
lo3 = widgets.HTML(value=f'<b>Signal 3: {signal_names[2]}</b>')
d3 = widgets.Dropdown(description='Direction',options=dir_options)
o3 = widgets.Dropdown(description='Order Type',options=order_options)
li3 = widgets.FloatText(description='Limit Price')
pt3 = widgets.FloatText(description='Profit Target')
pt3_opt = widgets.FloatRangeSlider(description='PT OPT',min=0,max=10.0,step=0.1)
pt3_step = widgets.FloatText(description='Step', value=1)
sl3 = widgets.FloatText(description='Stop Loss')
sl3_opt = widgets.FloatRangeSlider(description='SL OPT',min=0,max=10.0,step=0.1)
sl3_step = widgets.FloatText(description='Step', value=1)
c3 = widgets.FloatText(description='Cancel > N Bars')
c3_opt = widgets.FloatRangeSlider(description='Cancel OPT',min=0,max=10.0,step=0.1)
c3_step = widgets.FloatText(description='Step', value=1)
h3 = widgets.FloatText(description='Max Hold Bars')
h3_opt = widgets.FloatRangeSlider(description='Hold Opt',min=0,max=10.0,step=0.1)
h3_step = widgets.FloatText(description='Step', value=1)
q3 = widgets.FloatText(description='Quantity: ')
r3 = widgets.Checkbox(value=False, description='Reduce only')
v3 = widgets.Checkbox(value=False, description='Adjust for Volatility')
f3 = widgets.Checkbox(value=False, description='Force EOD')
opt3 = widgets.Checkbox(value=False,description='Run Optimized Backtest')
signal3_ui = widgets.VBox([l3,d3,o3,pt3,sl3,c3,h3,q3,r3,v3,f3])

##########


l4 = widgets.HTML(value=f'<b>Signal 4: {signal_names[3]}</b>')
lo4 = widgets.HTML(value=f'<b>Signal 4: {signal_names[3]}</b>')
d4 = widgets.Dropdown(description='Direction',options=dir_options)
o4 = widgets.Dropdown(description='Order Type',options=order_options)
li4 = widgets.FloatText(description='Limit Price')
pt4 = widgets.FloatText(description='Profit Target')
pt4_opt = widgets.FloatRangeSlider(description='PT OPT',min=0,max=10.0,step=0.1)
pt4_step = widgets.FloatText(description='Step', value=1)
sl4 = widgets.FloatText(description='Stop Loss')
sl4_opt = widgets.FloatRangeSlider(description='SL OPT',min=0,max=10.0,step=0.1)
sl4_step = widgets.FloatText(description='Step', value=1)
c4 = widgets.FloatText(description='Cancel > N Bars')
c4_opt = widgets.FloatRangeSlider(description='Cancel OPT',min=0,max=10.0,step=0.1)
c4_step = widgets.FloatText(description='Step', value=1)
h4 = widgets.FloatText(description='Max Hold Bars')
h4_opt = widgets.FloatRangeSlider(description='Hold Opt',min=0,max=10.0,step=0.1)
h4_step = widgets.FloatText(description='Step', value=1)
q4 = widgets.FloatText(description='Quantity: ')
r4 = widgets.Checkbox(value=False, description='Reduce only')
v4 = widgets.Checkbox(value=False, description='Adjust for Volatility')
f4 = widgets.Checkbox(value=False, description='Force EOD')
opt4 = widgets.Checkbox(value=False,description='Run Optimized Backtest')
signal4_ui = widgets.VBox([l4,d4,o4,pt4,sl4,c4,h4,q4,r4,v4,f4])

signal_line1 = widgets.HBox([signal1_ui,signal2_ui])
signal_line2 = widgets.HBox([signal3_ui,signal4_ui])

slip_comm = widgets.HTML(value='<b>Order Details</b>')
slip = widgets.FloatText(description='Slippage % per contract/share')
comm = widgets.FloatText(description='Commission $ per contract/share')
slip_comm_ui = widgets.HBox([slip,comm])
save_config = widgets.Checkbox(description='Save Config')

signal_acc = widgets.VBox([signal_line1,signal_line2])
order_acc = widgets.VBox([slip_comm,slip_comm_ui,save_config])


opt1_ui = widgets.VBox([lo1,opt1,pt1_opt,pt1_step,sl1_opt,sl1_step,c1_opt,c1_step,h1_opt,h1_step])
opt2_ui = widgets.VBox([lo2,opt2,pt2_opt,pt2_step,sl2_opt,sl2_step,c2_opt,c2_step,h2_opt,h2_step])
opt3_ui = widgets.VBox([lo3,opt3,pt3_opt,pt3_step,sl3_opt,sl3_step,c3_opt,c3_step,h3_opt,h3_step])
opt4_ui = widgets.VBox([lo4,opt4,pt4_opt,pt4_step,sl4_opt,sl4_step,c4_opt,c4_step,h4_opt,h4_step])
opt_line1 = widgets.HBox([opt1_ui, opt2_ui])
opt_line2 = widgets.HBox([opt3_ui, opt4_ui])
opt_ui = widgets.VBox([opt_line1,opt_line2])



# creating function to get values
def get_order_values(l1,li1,d1,o1,pt1,sl1,c1,h1,q1,r1,v1,f1,
                     l2,li2,d2,o2,pt2,sl2,c2,h2,q2,r2,v2,f2,
                     l3,li3,d3,o3,pt3,sl3,c3,h3,q3,r3,v3,f3,
                     l4,li4,d4,o4,pt4,sl4,c4,h4,q4,r4,v4,f4,
                     slip,comm, save_config):
    
    global ORDER_DF_REF
    dir_list = [d1,d2,d3,d4]
    order_list = [o1,o2,o3,o4]
    li_list = [li1,li2,li3,li4]
    pt_list = [pt1,pt2,pt3,pt4]
    sl_list = [sl1,sl2,sl3,sl4]
    c_list = [c1,c2,c3,c4]
    h_list = [h1,h2,h3,h4]
    q_list = [q1,q2,q3,q4]
    r_list = [r1,r2,r3,r4]
    v_list = [v1,v2,v3,v4]
    f_list = [f1,f2,f3,f4]
    s_list = [slip]*4
    comm_list = [comm]*4
    
    order_df = pd.DataFrame({'Signal':signal_names,
                           'Direction':dir_list,
                           'Order Type':order_list,
                            'Limit Price':li_list,
                           'Profit Target':pt_list,
                           'Stop Loss':sl_list,
                            'Cancel Order':c_list,
                             'Max Hold':h_list,
                           'Quantity':q_list,
                           'Reduce':r_list,
                           'Adjust Volatility':v_list,
                            'Force EOD':f_list,
                            'Slippage':s_list,
                            'Commission':comm_list})
    if save_config == True:
        order_df.to_csv(SIGNAL_DIR + 'config.csv')
    display(order_df)
    ORDER_DF_REF = order_df
    return order_df


def get_opt_values(pt1_opt,pt1_step,sl1_opt,sl1_step,c1_opt,c1_step,h1_opt,h1_step,
                    pt2_opt,pt2_step,sl2_opt,sl2_step,c2_opt,c2_step,h2_opt,h2_step,
                    pt3_opt,pt3_step,sl3_opt,sl3_step,c3_opt,c3_step,h3_opt,h3_step,
                    pt4_opt,pt4_step,sl4_opt,sl4_step,c4_opt,c4_step,h4_opt,h4_step,
                    opt1,opt2,opt3,opt4):
    
    global ORDER_DF_REF
    ORDER_DF_MOD = ORDER_DF_REF.copy()
    
    pt_list, sl_list, c_list, h_list = [],[],[],[]
    
    pt_opt_list = [pt1_opt,pt2_opt,pt3_opt, pt4_opt]
    sl_opt_list = [sl1_opt,sl2_opt,sl3_opt, sl4_opt]
    c_opt_list = [c1_opt,c2_opt,c3_opt, c4_opt]
    h_opt_list = [h1_opt,h2_opt,h3_opt, h4_opt]
    pt_step_list = [pt1_step,pt2_step,pt3_step,pt4_step]
    sl_step_list = [sl1_step,sl2_step,sl3_step,sl4_step]
    c_step_list = [c1_step,c2_step,c3_step,c4_step]
    h_step_list = [h1_step,h2_step,h3_step,h4_step]
    opt_list = [opt1,opt2,opt3,opt4]
    
    pt_list = [list(np.arange(pt1_opt[0],pt1_opt[1], pt1_step)),
               list(np.arange(pt2_opt[0],pt2_opt[1], pt2_step)),
               list(np.arange(pt3_opt[0],pt3_opt[1], pt3_step)),
               list(np.arange(pt4_opt[0],pt4_opt[1], pt4_step))]
    
    sl_list = [list(np.arange(sl1_opt[0],sl1_opt[1], sl1_step)),
               list(np.arange(sl2_opt[0],sl2_opt[1], sl2_step)),
               list(np.arange(sl3_opt[0],sl3_opt[1], sl3_step)),
               list(np.arange(sl4_opt[0],sl4_opt[1], sl4_step))]
    
    c_list = [list(np.arange(c1_opt[0],c1_opt[1], c1_step)),
               list(np.arange(c2_opt[0],c2_opt[1], c2_step)),
               list(np.arange(c3_opt[0],c3_opt[1], c3_step)),
               list(np.arange(c4_opt[0],c4_opt[1], c4_step))]
    
    h_list = [list(np.arange(h1_opt[0],h1_opt[1], h1_step)),
               list(np.arange(h2_opt[0],h2_opt[1], h2_step)),
               list(np.arange(h3_opt[0],h3_opt[1], h3_step)),
               list(np.arange(h4_opt[0],h4_opt[1], h4_step))]

    
    ORDER_DF_REF['Optimize'] = opt_list
    ORDER_DF_MOD['PTOpt'] = pt_list
    ORDER_DF_MOD['SLOpt'] = sl_list
    ORDER_DF_MOD['COpt'] = c_list
    ORDER_DF_MOD['HOpt'] = h_list
    
    ORDER_DF_REF['Profit Target'] = [p1 if o==False else p2 for (p1,o,p2) in zip(ORDER_DF_REF['Profit Target'],ORDER_DF_REF['Optimize'],ORDER_DF_MOD['PTOpt'])]
    ORDER_DF_REF['Stop Loss'] = [s1 if o==False else s2 for (s1,o,s2) in zip(ORDER_DF_REF['Stop Loss'],ORDER_DF_REF['Optimize'],ORDER_DF_MOD['SLOpt'])]
    ORDER_DF_REF['Cancel Order'] = [c1 if o==False else c2 for (c1,o,c2) in zip(ORDER_DF_REF['Cancel Order'],ORDER_DF_REF['Optimize'],ORDER_DF_MOD['COpt'])]
    ORDER_DF_REF['Max Hold'] = [h1 if o==False else h2 for (h1,o,h2) in zip(ORDER_DF_REF['Max Hold'],ORDER_DF_REF['Optimize'],ORDER_DF_MOD['HOpt'])]
    
    display(ORDER_DF_REF)
    ORDER_DF_REF = ORDER_DF_REF
    return ORDER_DF_REF

    

ORDER_DF = interactive_output(get_order_values,
                              {'l1':l1,'li1':li1,'d1':d1,'o1':o1,'pt1':pt1,'sl1':sl1,'r1':r1,'v1':v1,'q1':q1,'c1':c1,'h1':h1,'f1':f1,
                               'l2':l2,'li2':li2,'d2':d2,'o2':o2,'pt2':pt2,'sl2':sl2,'r2':r1,'v2':v2,'q2':q2,'c2':c2,'h2':h2,'f2':f2,
                               'l3':l3,'li3':li3,'d3':d3,'o3':o3,'pt3':pt3,'sl3':sl3,'r3':r1,'v3':v3,'q3':q3,'c3':c3,'h3':h3,'f3':f3,
                               'l4':l4,'li4':li4,'d4':d4,'o4':o4,'pt4':pt4,'sl4':sl4,'r4':r1,'v4':v4,'q4':q4,'c4':c4,'h4':h4,'f4':f4,
                               'slip':slip,'comm':comm, 'save_config':save_config})

OPT = interactive_output(get_opt_values,
                        {'pt1_opt':pt1_opt,'sl1_opt':sl1_opt,'c1_opt':c1_opt,'h1_opt':h1_opt,'pt1_step':pt1_step,'sl1_step':sl1_step,'c1_step':c1_step,'h1_step':h1_step,
                        'pt2_opt':pt2_opt,'sl2_opt':sl2_opt,'c2_opt':c2_opt,'h2_opt':h2_opt,'pt2_step':pt2_step,'sl2_step':sl2_step,'c2_step':c2_step,'h2_step':h2_step,
                        'pt3_opt':pt3_opt,'sl3_opt':sl3_opt,'c3_opt':c3_opt,'h3_opt':h3_opt,'pt3_step':pt3_step,'sl3_step':sl3_step,'c3_step':c3_step,'h3_step':h3_step,
                        'pt4_opt':pt4_opt,'sl4_opt':sl4_opt,'c4_opt':c4_opt,'h4_opt':h4_opt,'pt4_step':pt4_step,'sl4_step':sl4_step,'c4_step':c4_step,'h4_step':h4_step,
                        'opt1':opt1,'opt2':opt2,'opt3':opt3,'opt4':opt4})

accordion = widgets.Accordion(children=[signal_acc,opt_ui,order_acc,OPT])
accordion.set_title(0, 'Signal Details')
accordion.set_title(1, 'Optimization')
accordion.set_title(2, 'Order Details')
accordion.set_title(3, 'Confirm Options')

# creating master UI
ORDER_UI = accordion
```
